```python
from pyspark import SparkConf, SparkContext
import json
import sys
## Constants
APP_NAME = "Json to CSV"
##OTHER FUNCTIONS/CLASSES

def process(tweet):
    json_obj = None
    try:
        json_obj = json.loads(tweet.replace('#',""))
    except:
        #print "Unexpected error:", sys.exc_info()[0]
        return None
    if 'delete' in json_obj.keys():
        return None
    ls = []
    for i in range(0,len(json_obj['entities']['hashtags'])):
        try:
            #print 'writing'
            text = str(str(json_obj['text'])).replace(",","")
            hashtag = str(json_obj['entities']['hashtags'][i]['text']).replace(",","")
            ls.append(text+','+hashtag)
        except:
            #print "Unexpected error:", sys.exc_info()[0]
            continue
    #master_csv.flush()
    return ls

def main(sc):
    datafile = sc.textFile("hdfs://master:9000/data/2015/03/22/*/*.json")
    temp = datafile.map(lambda x: process(x))
    csv = temp.filter(lambda x: x!=None and len(x)!=0).flatMap(lambda x: x)
    csv.saveAsTextFile("hdfs://master:9000/data/csv/")

if __name__ == "__main__":
    conf = SparkConf().setAppName(APP_NAME)
    conf = conf.setMaster("spark://master:7077")
    conf = conf.set("spark.submit.deployMode", "cluster")
    sc= SparkContext(conf=conf)
    filename = sys.argv[1]
    # Execute Main functionality
    main(sc)
```
